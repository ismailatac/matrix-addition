```python
import torch
import time
import matplotlib.pyplot as plt

# Matris boyutları
matrix_sizes = [5000, 10000, 15000, 20000, 25000]

# Matris toplama işlemi ve süre kaydı
timings = []
for size in matrix_sizes:
    A = torch.randn(size, size)
    B = torch.randn(size, size)

    start_time = time.time()
    C = A + B
    end_time = time.time()

    timings.append(end_time - start_time)
    print("Matris boyutu: ", size)
    print("Çalışma süresi: ", end_time - start_time, "saniye")
    speedup = [timings[0] / t for t in timings]
    print("Speedup: ", speedup[-1])
    efficiency = [s / size for s, size in zip(speedup, matrix_sizes)]
    print("Efficiency: ",efficiency[-1])
    print("---------------------------")




# Zaman grafiği
plt.plot(matrix_sizes, timings, label="Zaman")
plt.xlabel("Matris boyutu")
plt.ylabel("Çalışma süresi (saniye)")
plt.legend()
plt.show()

# Speedup grafiği
speedup = [timings[0] / t for t in timings]
plt.plot(matrix_sizes, speedup, label="Speedup")
plt.xlabel("Matris boyutu")
plt.ylabel("Speedup")
plt.legend()
plt.show()

# Efficiency grafiği
efficiency = [s / size for s, size in zip(speedup, matrix_sizes)]
plt.plot(matrix_sizes, efficiency, label="Efficiency")
plt.xlabel("Matris boyutu")
plt.ylabel("Efficiency")
plt.legend()
plt.show()
```
